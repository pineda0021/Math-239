## **Part 2: Plot the histogram for the EWL (y) and conduct the normality tests**
---


🧑‍💻🏴‍☠️🦜 **RStudio**

```python
%%R
library(ggplot2)

ggplot(weightloss.data, aes(x = EWL)) +
  geom_histogram(aes(y = ..density..), bins = 9, fill = "blue", color = "black") +
  stat_function(fun = dnorm, args = list(mean = mean(weightloss.data$EWL), sd = sd(weightloss.data$EWL)), color = "red", size = 1.2) +
  labs(title = "Weight Loss Histogram with Normal Distribution",
       x = "EWL",
       y = "Density") +
  theme_bw()
```

```python
%%R
shapiro.test(EWL)
```

🧑‍💻🇵‌🇾‌🐍 **Python**

```python
import plotly.express as px

# Create a histogram
fig = px.histogram(df, x="EWL",
                   title="Weight Loss Histogram with Normal Distribution")
fig.show()
```

```python
from scipy.stats import shapiro

# Perform the Shapiro-Wilk test
shapiro_test = shapiro(df["EWL"])

print(shapiro_test)

# Interpret the results
alpha = 0.05  # Significance level
if shapiro_test.pvalue > alpha:
    print("The data is likely normally distributed (fail to reject H0)")
else:
    print("The data is likely not normally distributed (reject H0)")
```

**Based on the histograms from R and Python, the data appears to follow a normal distribution. Additionally, the Shapiro-Wilk normality test shows a p-value greater than the alpha level of 0.05, indicating that the response variable is normally distributed.**

## **Part 3: Fit a general linear model**
---

🧑‍💻🏴‍☠️🦜 **RStudio**

```python
%%R
# Ensure 'drug' and 'gender' are factors
weightloss.data$drug <- as.factor(weightloss.data$drug)
weightloss.data$gender <- as.factor(weightloss.data$gender)

# Relevel the factors directly within the data frame
drug.rel <- relevel(weightloss.data$drug, ref = "A")
gender.rel <- relevel(weightloss.data$gender, ref = "M")

# Fit the model using the modified factors
fitted.model <- glm(EWL ~ drug.rel + age + gender.rel, data = weightloss.data, family=gaussian(link=identity))

# Display the summary of the model
summary(fitted.model)
```

🧑‍💻🇵‌🇾‌🐍 **Python**

```python
import statsmodels.formula.api as sm

# Convert relevant columns to categorical
df['drug'] = pd.Categorical(df['drug'], categories=['A', 'B'], ordered=False)
df['gender'] = pd.Categorical(df['gender'], categories=['M', 'F'], ordered=False)

# Fit the model
model = sm.ols('EWL ~ C(drug, Treatment(reference="A")) + age + C(gender, Treatment(reference="M"))', data=df).fit()

# Display the summary
print(model.summary())
```

**(b)** Which regression cofficients turn out to be significant at the 5%?

**Drug B is the only significant predictor in the model at the 5% significance level since its p-value is less than the 0.05 alpha level.**


**(c)** Is one of the drugs more efficient for weight loss than the other? Interpret all estimated signifficant coefficients.

**The estimated average EWL for subjects taking drug B is 4.8103 percent higher than that for subjects taking drug A, keeping all the other predictors fixed. It means that drug B is more efficient than drug A.**

## **Part 4: Model Fits the data well ?**
---


🧑‍💻🏴‍☠️🦜 **RStudio**

```python
%%R
#outputting estimated sigma
sigma(fitted.model)
```

```python
%%R
#checking model fit
null.model<- glm(EWL ~ 1, data=weightloss.data, family=gaussian(link=identity))
print(deviance<- -2*(logLik(null.model)-logLik(fitted.model)))
print(p.value<- pchisq(deviance, df=3, lower.tail=FALSE))
```

🧑‍💻🇵‌🇾‌🐍 **Python**

```python
import statsmodels.api as sm
from scipy import stats

# Calculate sigma (standard deviation of the residuals)
sigma = model.resid.std()

print(f"Sigma (standard deviation of residuals): {sigma}")

# Checking model fit
null_model = sm.OLS.from_formula("EWL ~ 1", data=df).fit()

deviance = -2 * (null_model.llf - model.llf)
print(f"Deviance: {deviance}")

# p-value calculation using chi-squared distribution
# Use stats.chi2.sf instead of the deprecated chisqprob
p_value = stats.chi2.sf(deviance, 3)
print(f"P-value: {p_value}")
```

**(b)** Discuss goodness-of-it of the model.

**The p-value for the deviance test is less than 0.05, indicating a good fit for the model.**

**(a)** Fit a general linear regression model to predict the price of a car. Write down the fitted model, specifying all estimated parameters.

🧑‍💻🏴‍☠️🦜 **RStudio**

```python
%%R
#rescaling price by a factor of 1000
priceK <- price / 1000
#specifying reference levels
# Convert variables to factors before applying relevel
carsales.data$bodystyle <- as.factor(carsales.data$bodystyle)
carsales.data$country <- as.factor(carsales.data$country)
carsales.data$leather <- as.factor(carsales.data$leather)

bodystyle.rel<- relevel(carsales.data$bodystyle, ref="hatchback")
country.rel<- relevel(carsales.data$country, ref="Japan")
leather.rel<- relevel(carsales.data$leather, ref="no")
#fitting general linear model
summary(fitted.model<- glm(priceK ~ bodystyle.rel + country.rel + hwy + doors
+ leather.rel, data=carsales.data, family=gaussian(link=identity)))
```

```python
%%R
#outputting estimated sigma
sigma(fitted.model)
```

🧑‍💻🇵‌🇾‌🐍 **Python**

```python
import statsmodels.formula.api as smf
import pandas as pd


# Convert relevant columns to categorical
df['bodystyle'] = pd.Categorical(df['bodystyle'], categories=df['bodystyle'].unique())
df['country'] = pd.Categorical(df['country'], categories=df['country'].unique())
df['leather'] = pd.Categorical(df['leather'], categories=df['leather'].unique())

# Set reference levels
df['bodystyle'] = df['bodystyle'].cat.reorder_categories(['hatchback'] + list(cat for cat in df['bodystyle'].cat.categories if cat != 'hatchback'))
df['country'] = df['country'].cat.reorder_categories(['Japan'] + list(cat for cat in df['country'].cat.categories if cat != 'Japan'))
df['leather'] = df['leather'].cat.reorder_categories(['no'] + list(cat for cat in df['leather'].cat.categories if cat != 'no'))

# Fit the general linear model using the formula API
# Note the change from sm.ols to smf.ols
model = smf.ols('priceK ~ C(bodystyle) + C(country) + hwy + doors + C(leather)', data=df).fit()

# Print the model summary
print(model.summary())
```

```python
import statsmodels.api as sm
from scipy import stats


# Calculate sigma (standard deviation of the residuals)
sigma = model.resid.std()

print(f"Sigma (standard deviation of residuals): {sigma}")
```

**The fitted model is**  $\hat{E}\left(priceK\right)=5.1353+2.2698\cdot coupe+6.4107\cdot sedan+3.1959\cdot Germany+3.2128\cdot USA+0.1305\cdot hwy+1.5554\cdot doors+12.1757\cdot leather,$  for R: $\hat{\sigma}=3.48$, and
Python: $\hat{\sigma}=2.98$.
